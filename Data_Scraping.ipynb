```python
import ScraperFC.sofascore as sfc
import pandas as pd

ss = sfc.Sofascore()

season = "25/26"

leagues = [
    "EPL",                  # Premier League
    "Bundesliga",           # Bundesliga
    "La Liga",               # LaLiga
    "Serie A",              # Serie A
    "Ligue 1",              # Ligue 1
    "Champions League",
    "UEFA Nations League"   # National teams (UEFA)
]

all_matches = []

for league in leagues:
    try:
        matches = ss.get_match_dicts(season, league)

        for m in matches:
            match_data = {
                "id": m.get("id"),
                "tournament": league,
                "season": season,
                "hometeam": m.get("homeTeam", {}).get("name"),
                "awayteam": m.get("awayTeam", {}).get("name"),
                "homescore": m.get("homeScore", {}).get("current"),
                "awayscore": m.get("awayScore", {}).get("current"),
                "hasGlobalHighlights": m.get("hasGlobalHighlights"),
                "hasXg": m.get("hasXg"),
                "hasEventPlayerStatistics": m.get("hasEventPlayerStatistics"),
                "hasEventPlayerHeatMap": m.get("hasEventPlayerHeatMap"),
            }
            all_matches.append(match_data)

        print(f"Done: {league}")

    except Exception as e:
        print(f"Failed: {league} -> {e}")

df = pd.DataFrame(all_matches)
df.to_csv("matches.csv", index=False)

print("Saved to matches.csv")
```
