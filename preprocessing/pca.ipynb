```python
import sklearn.decomposition as skd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN, HDBSCAN
from sklearn.neighbors import KNeighborsClassifier
```

```python
def train_pca(path_to_csv):
    data_matrix = pd.read_csv(path_to_csv).values
    pca = skd.PCA(n_components = 'mle', svd_solver = 'full') # Minka’s MLE is used to guess the dimension.
    pca.fit(data_matrix)
    
    print("componets:", pca.n_components_)
    print("Explained variance ratio per pc:", pca.explained_variance_ratio_)

    return pca

def apply_pca(pca, path_to_csv):
    data_matrix = pd.read_csv(path_to_csv).values
    return pca.transform(data_matrix)
```

```python
def train_clustering(data):
    clustering = HDBSCAN(min_cluster_size=11)
    clustering.fit(data)
    return clustering

def plot_clustering(clustering, data):
    plt.scatter(data[:, 0], data[:, 1], c=clustering.labels_, cmap='viridis')
    plt.title('HDBSCAN Clustering')
    plt.xlabel('Principal Component 1')
    plt.ylabel('Principal Component 2')
    plt.colorbar(label='Cluster Label')
    plt.show()
    
def predict_cluster_membership(base_data, clustering, to_predict):
    #filter out noise first
    mask = clustering.labels_ != -1
    clean_data = base_data[mask]
    clean_labels = clustering.labels_[mask]
    
    # Fit KNN only on valid clusters, then predict the cluster for the new data point based on nearest neighbors
    knn = KNeighborsClassifier(n_neighbors=1) 
    knn.fit(clean_data, clean_labels)
    return knn.predict(to_predict)
    
```
