# Data Preparation


```python
# find matchpaths
import json
import os
START_DATE = 2010

# find all competitions
competition_paths = []
with open("..\\data\\competitions.json", "r") as f:
    competitions = json.load(f)
for competition in competitions:
    if competition["competition_gender"] == "male" and int(competition["season_name"].split("/")[0]) >= START_DATE:
        competition_paths.append(os.path.join("..", "data", "matches", str(competition["competition_id"]), str(competition["season_id"]) + ".json"))

print(f"Found {len(competition_paths)} competitions")

# find all matches
matches_paths = []
for path in competition_paths:
    with open(path, "r") as f:
        matches = json.load(f)
        for match in matches:
            matches_paths.append(os.path.join("..", "data", "events", str(match["match_id"]) + ".json"))
    print(f"Found {len(matches)} matches in {path}")
print(f"Found {len(matches_paths)} matches in total")
```
